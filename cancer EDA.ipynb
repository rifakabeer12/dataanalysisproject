# Exploratory Data Analysis

This project aims to model and predict the presence of breast cancer. This dataset of breast cancer patients was obtained from the 2017 November update of the SEER Program of the NCI, which provides information on population-based cancer statistics.

<font color="RED">Data preprocessing</font>


```python
# check the presence of missing value

data.isnull().sum()
```

```python
# handling categorical data

X= pd.get_dummies(data,drop_first=True)
X.head()
```

```python
Y=pd.get_dummies(data['Tumor Size'],drop_first=True)
Y.head()
```

```python
# split the dataset

from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test= train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

<font color="RED">Feature Scaling:</font>


Before applying algorithm we should check whether the data is equally splitted or not, because if data is not splitted equally it will cause for data imbalacing problem

```python
# standardization:
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# normalisation
from sklearn.preprocessing import MinMaxScaler
mmsc= MinMaxScaler()
x_train = mmsc.fit_transform(x_train)
x_test = mmsc.transform(x_test)
```
